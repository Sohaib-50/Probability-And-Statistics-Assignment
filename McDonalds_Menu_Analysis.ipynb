importing necessary libraries for analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
```

Reading the McDonald's menu data set and displaying initial few rows

```python
dataset = pd.read_csv('McDonalds Menu.csv')  # read the data set
dataset.head(10)  # show first few rows
```

```python
# show all unique categories
dataset['Category'].unique()
```

```python
# split data set into drinks and non-drinks where drinks is the data set with Category = 'Beverages', 'Coffee & Tea', or 'Smoothies & Shakes'
drinks = dataset[dataset['Category'].isin(['Beverages', 'Coffee & Tea', 'Smoothies & Shakes'])]
non_drinks = dataset[~dataset['Category'].isin(['Beverages', 'Coffee & Tea', 'Smoothies & Shakes'])]
```

```python
# show random sample of drinks
drinks.sample(5)
```

```python
# show random sample of non-drinks
non_drinks.sample(5)
```

Now extracting and setting grams value for Serving Size of each non drink item

```python
non_drinks['Serving Size'] = non_drinks['Serving Size'].str.split('(', expand=True)[1].str.split('g', expand=True)[0].astype(float)
```

```python
# show Serving Size column
non_drinks['Serving Size']
```

Adding a column for calorie density of each non drink item

```python
# add a new column for calorie density of each non-drink item
non_drinks['Calorie Density'] = non_drinks['Calories'] / non_drinks['Serving Size']
```

```python
# show calorie density column
non_drinks['Calorie Density']
```

```python
# Show the top 10 items by calorie density
non_drinks.sort_values(by='Calorie Density', ascending=False).head(10)
```

```python
# show the bottom 10 items by calorie density
non_drinks.sort_values(by='Calorie Density', ascending=True).head(10)
```

```python
# show Item column for top 3 items by calorie density for each Category in non-drinks sorted by Category
most_calorie_dense = non_drinks.sort_values(by=["Category", 'Calorie Density'], ascending=[True, False]).groupby('Category')[["Item", "Calorie Density", "Category"]].head(3)
most_calorie_dense
```

```python
# show Item column for top 3 items by calorie density for each Category in non-drinks sorted by Category
non_drinks.sort_values(by=["Category", 'Calorie Density'], ascending=[True, True]).groupby('Category')[["Item", "Calorie Density", "Category"]].head(3)
```

```python
# plot most_calorie_dense Item, category and calorie density with each Category bar colored according to Category
# most_calorie_dense.plot(kind='bar', x='Category', y='Calorie Density', color='Category')
pivoted = most_calorie_dense.pivot(index='Item', columns='Category', values='Calorie Density')
ax = pivoted.plot(kind='barh', stacked=True) 
ax.legend(title='Attribute', bbox_to_anchor=(1, 1), loc='upper left')

#  x = most_calorie_dense.plot(kind='barh', title='Bar Chart of Colors', figsize=(6, 4))
# x.set(xlabel='Items')
# x.legend(title='Attribute', bbox_to_anchor=(1, 1), loc='upper left')
# x.grid(axis='x')

```

```python
# create a new column for calories in 300g serving size
non_drinks['Calories in 300g'] = non_drinks['Calorie Density'] * 300
```

```python
non_drinks["Servings for 300g"] = 300 / non_drinks['Serving Size']
```

```python
# show the item with least value of servings for 500g
non_drinks.sort_values(by='Servings for 500g', ascending=True).head(1)
```
