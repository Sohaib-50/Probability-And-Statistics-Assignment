```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('McDonalds Menu.csv')  # read the data set
```

```python
# seperate out grilled chicken and crispy chicken items
grilled_chicken_items = dataset[dataset['Item'].str.contains('Grilled')]
crispy_chicken_items = dataset[dataset['Item'].str.contains('Crispy')]
```

```python
grilled_chicken_items.sample(5)["Item"]
```

```python
crispy_chicken_items.sample(5)["Item"]
```

```python
# make a box plot of calories of crispy chicken items vs grilled chicken items
plt.boxplot([grilled_chicken_items['Calories'], crispy_chicken_items['Calories']])
# label the boxes
plt.xticks([1, 2], ['Grilled Chicken', 'Crispy Chicken'])
plt.title('Calories of Grilled and Crispy Chicken items')
plt.show()
```

```python
comparator = "Total Fat"
plt.boxplot([grilled_chicken_items[comparator], crispy_chicken_items[comparator]])
plt.xticks([1, 2], ['Grilled Chicken', 'Crispy Chicken'])
plt.title(f'{comparator} of Grilled and Crispy Chicken items')
plt.show()
```

```python
comparator = "Protein"
plt.boxplot([grilled_chicken_items[comparator], crispy_chicken_items[comparator]])
plt.xticks([1, 2], ['Grilled Chicken', 'Crispy Chicken'])
plt.title(f'{comparator} of Grilled and Crispy Chicken items')
plt.show()
```

```python

# boxplot of sum of Total Fat, Calories, Saturated Fat, Cholestrol, and Sugars of grilled and crispy chicken items
plt.boxplot([grilled_chicken_items[["Total Fat", "Calories", "Saturated Fat", "Cholesterol", "Sugars"]].sum(axis=1),
            crispy_chicken_items[["Total Fat", "Calories", "Saturated Fat", "Cholesterol", "Sugars"]].sum(axis=1)])
plt.xticks([1, 2], ['Grilled Chicken', 'Crispy Chicken'])
plt.title(f'Sum of Total Fat, Calories, Saturated Fat, Cholestrol, and Sugars of grilled and crispy chicken items')
plt.show()
```

```python
# boxplot of sum of Dietary Fiber and Protein
plt.boxplot([grilled_chicken_items[["Dietary Fiber", "Protein"]].sum(axis=1),
            crispy_chicken_items[["Dietary Fiber", "Protein"]].sum(axis=1)])
plt.xticks([1, 2], ['Grilled Chicken', 'Crispy Chicken'])
plt.title(f'Sum of Dietary Fiber and Protein  of grilled and crispy chicken items')
plt.show()
```
